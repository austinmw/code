iv. Create a new qualitative variable, called Elite, by binning the Top10perc variable.  
We are going to divide universities into two groups based on whether or not the proportion of students coming from the top 10% of their high school classses exceeds 50%.  

```r
Elite = rep("No",nrow(college)) # rep replicates the values in first arg
Elite[college$Top10perc > 50]="Yes"
Elite = as.factor(Elite)
college = data.frame(college,Elite) # join Elite to college as new column
head(college,3)
```

Use the summary function to see how many elite universities there are. Now use the plot() function to produce side-by-side boxplots of Outstate versus Elite.

```r
summary(college$Elite)
plot(college$Elite,college$Outstate,xlab='Elite',ylab='Outstate')
```

```r
par(mfrow=c(2,2)) # divide print window into regions
hist(college$Apps)
hist(college$perc.alumni, col=2) #col=2 means color red
hist(college$S.F.Ratio, col=3, breaks=10)
hist(college$Expend, breaks=100)
```

vi. Continue exploring the data, and provide a brief summary of what you discover.

```r
# High tuition correlates to high graduation rate.
plot(college$Outstate, college$Grad.Rate)
# Colleges with low acceptance rate tend to have low S:F ratio.
plot(college$Accept / college$Apps, college$S.F.Ratio)
# Colleges with the most students from top 10% perc don't necessarily have
# the highest graduation rate. Also, rate > 100 is erroneous!
plot(college$Top10perc, college$Grad.Rate)
```

**(c)** Are any of the predictors associated with per capita crime rate? If so, explain the relationship.

```r
# Older homes, more crime
plot(Boston$age, Boston$crim)
# Closer to work-area, more crime
plot(Boston$dis, Boston$crim)
# Higher index of accessibility to radial highways, more crime
plot(Boston$rad, Boston$crim)
# Higher tax rate, more crime
plot(Boston$tax, Boston$crim)
# Higher pupil:teacher ratio, more crime
plot(Boston$ptratio, Boston$crim)
```

**(d)** Do any of the suburbs of Boston appear to have particularly high crime rates? Tax rates? Pupil-teacher ratios? Comment on the range of each predictor.

```r
par(mfrow=c(1,3))
# most cities have low crime rates, but there is a long tail: 18 suburbs appear
# to have a crime rate > 20, reaching to above 80
hist(Boston$crim[Boston$crim>1], breaks=25) # only elements > 1.00
# there is a large divide between suburbs with low tax rates and a peak at 660-680
hist(Boston$tax, breaks=25)
# a skew towards high ratios, but no particularly high ratios
hist(Boston$ptratio, breaks=25)
```
