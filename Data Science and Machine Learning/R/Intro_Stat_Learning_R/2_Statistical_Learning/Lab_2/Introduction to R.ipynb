```r
# random normal variables, with first argument n the sample size
x=rnorm(50) # by default, mean 0, std 1
y=x+rnorm(50,mean=50,sd=.1)
cor(x,y)
```

```r
# mean() and var() of a vector of numbers
# applying sqrt() to var() will give the std, or can just use sd()
set.seed(3)
y=rnorm(100)
mean(y)
var(y)
sqrt(var(y))
sd(y)
```

**2.3.2 Graphics**

```r
# plot() function
x=rnorm(100)
y=rnorm(100)
plot(x,y) # produces scatterplot
# xlab, ylab, main
plot(x,y,xlab="this is the x-axis", ylab="this is the y-axis", main="Plot of X vs Y")
```

```r
# save the output of an R plot
pdf("test_files/test_pdf.pdf")
plot(x,y,col="green")
# the dev.off() function indicates that we are done creating the plot
dev.off()
```

```r
# create a sequence of numbers
x=seq(1,10)
x
x=1:10
x
x=seq(-pi,pi,length=5)
x
```

```r
# contour plot to represent three-dimensional data (topographical map)
# takes min of 3 args: vector of x vals, vector of y vals, matrix of z vals
y=x
f=outer(x,y,function(x,y)cos(y)/(1+x^2)) # outer product of arrays
contour(x,y,f)
contour(x,y,f,nlevels=45,add=T)
fa=(f-t(f))/2
contour(x,y,fa,nlevels=15)
```

```r
# image() works same as contour, except that it produces a color-coded plot
# color depends on z value, known as a heatmap
image(x,y,fa)
# persp() can be used to produce a three-dimensional plot
# theta and phi control the angles at which the plot is viewed
persp(x,y,fa)
persp(x,y,fa,theta=30)
persp(x,y,fa,theta=30,phi=20)
persp(x,y,fa,theta=30,phi=70)
persp(x,y,fa,theta=30,phi=40)
```

**2.3.4 Loading Data**

```r
# read.table(), write.table() can be used to import/export data
# must tell R to search for the data in the proper directory
# begin by loading the Auto data set, part of the ISLR library
# load Auto.data file into R and store it as Auto object (data frame format)
Auto=read.table("Auto.data")
# fix() can be used to view data in a spreadsheet like window
# fix(Auto)
# doesn't seem to work in jupyter
Auto
```

```r
# option header=T tells R that first line of file contains variable names
# option na.strings tells R that any time it sees particular chars,
# it should be treated as a missing element of the data matrix
# shows as ? before using na.strings option
Auto=read.table("Auto.data",header=T)
Auto[33,]
# shows as NA after na.strings option
Auto=read.table("Auto.data",header=T,na.strings="?")
Auto[33,]
```

```r
# CSV
Auto=read.csv("Auto.csv", header=T,na.strings="?")
head(Auto)
dim(Auto)
Auto[1:4,]
```

```r
# omit rows containing missing observations
Auto=na.omit(Auto)
dim(Auto) # 5 rows removed
```

**2.3.5 Additional Graphical and Numerical Summaries**

```r
# refer to variables in data sets using $ to join
plot(Auto$cylinders, Auto$mpg)
# attach() makes the variables in this data frame available by name
attach(Auto)
plot(cylinders, mpg)
```

```r
# convert quantitative variables into qualitative variables
# (cylinders variable is stored as a numeric vector [quantitative], 
# since there's only a small number of possible values for cylinders, may
# prefer to treat it as a qualitative variable)
cylinders
cylinders=as.factor(cylinders)
cylinders
```

```r
# If x-axis var is categorical, then boxplots will automatically be 
# produced by the plot() function
plot(cylinders, mpg)
plot(cylinders, mpg, col="red")
plot(cylinders, mpg, col="red", varwidth=T) # widths proportional to the square-roots of the number of observations in the groups
#?boxplot
plot(cylinders, mpg, col="red", varwidth=T, horizontal=T)
plot(cylinders, mpg, col="red", varwidth=T, xlab="cylinders",ylab="MPG")
```

```r
# histogram
hist(mpg)
hist(mpg,col=2) # col=2 has the same effect as col="red"
hist(mpg,col=2,breaks=15)
```

```r
# scatterplot matrix with pairs() 
# (scatterplot for every pair of vars for a given data set)
pairs(Auto)
# scatterplots for a subset of the variables
pairs(~ mpg + displacement + horsepower + weight + acceleration, Auto)
```

```r
# identify() provides a useful interactive method for identifying
# the value for a particular variable for points on a plot
# click to print value, control+click to exit identify() function
plot(horsepower, mpg)
# x-axis var, y-axis var, var whose values we would like to see printed
identify(horsepower, mpg, name) # doesn't seem to work in jupyter
```

```r
# summary of each variable in a data set
summary(Auto)
```
